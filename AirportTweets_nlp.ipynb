```
from nltk import word_tokenize, pos_tag
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt_tab') #tokenization
nltk.download('wordnet') #lemmatization

lemmatizer = WordNetLemmatizer()

#POS tag converter
#Part-Of-Speech tagging is the process of labeling each word in a sentence with its grammatical role, important for lemmatization to reduce a word to its base or dictionary form
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  #default

def preprocess(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", '', text) #removing urls
    text = re.sub(r'@\w+|\#', '', text) #removing mentions and hashtags
    text = re.sub(r"&amp;", "and", text)  #replacing HTML amp with "and"
    text = re.sub(r"[^a-zA-Z\s]", '', text) #removing punctuation and special chars (sadece letter ve space kalıyor)
    text = re.sub(r"\s+", ' ', text).strip() #removing whitespace

    tokens = word_tokenize(text)
    tagged_tokens = pos_tag(tokens)

    clean_tokens = [
        lemmatizer.lemmatize(word, get_wordnet_pos(tag))
        for word, tag in tagged_tokens
        if word not in stop_words and len(word) > 2
    ]

    return " ".join(clean_tokens)

df['clean_tweet'] = df['tweet'].apply(preprocess)

```

```
#encoding categorical labels into int

label_dict = {'negative': 0, 'neutral': 1, 'positive': 2}
df['label_id'] = df['label'].map(label_dict)

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

#custom for BERT input format
class TweetDataset(Dataset):
    def __init__(self, texts, labels):
        self.encodings = tokenizer(list(texts), truncation=True, padding=True, max_length=128)
        self.labels = labels

    def __len__(self): return len(self.labels)
    def __getitem__(self, idx):
        #returns the input text and corresponding label as tensors
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

#tensor is a multi-dimensional array similar optimized for GPU computation
```

```
from sklearn.model_selection import train_test_split

#Initial split into train+val and test
X_temp, X_test, y_temp, y_test = train_test_split(df['tweet'], df['label_id'], test_size=0.2, stratify=df['label_id'], random_state=42) #fixing the seed of the random number generator
#the operation will produce the same result every time

#stratify ensures the train-test splits maintain the same class proportions as the original dataset, preventing imbalance issues during model training and eval

#Further split train+val into train and validation
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1, stratify=y_temp, random_state=42)

#Converting to datasets
train_dataset = TweetDataset(X_train.tolist(), y_train.tolist())
val_dataset = TweetDataset(X_val.tolist(), y_val.tolist())
test_dataset = TweetDataset(X_test.tolist(), y_test.tolist())

model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=3) #model architecture from the Hugging Face Transformers library
#reads text in both directions to understand the context better

training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=64,
    warmup_steps=50, #used to gradually increase the learning rate from a small value to a larger one. This is done to prevent large updates early in training that might lead to instability
    eval_strategy="epoch", #after every epoch
    logging_dir='./logs',
    logging_steps=10,
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="accuracy", #selects the best model according to accuracy
    report_to="none" #for wandb

)

#Batch size refers to how many training examples are processed before the model's internal parameters are updated
#Larger: faster training but require more memory

#warmup_steps=50 means the learning rate will start small and will gradually increase over the first 50 training steps
#augmentation ya da early stopping yapmaya gerek olmadı
```

```
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

#precision: TP/(TP+FP) positive predictionlardan kaç tanesi doğru
#recall: TP/(TP+FN) model actual positivelerden kaç tanesini doğru tahmin edebildi
#F1 score: 2* Precision*Recall / (Precision+Recall)

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1) #gets predicted class by finding max logit
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='weighted')
    acc = accuracy_score(labels, preds)
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }
#logits are the raw, unnormalized scores output by a model before applying an activation function like softmax to convert them into probs

#fine-tuning
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    #eval_dataset=test_dataset,
    eval_dataset=val_dataset, #training'de görülmeyen verileri de deniyor
    compute_metrics=compute_metrics
)

trainer.train()
results = trainer.evaluate()
print("BERT Accuracy:", results['eval_accuracy'])

#I did two steps training to see better results
#This part evaluates the model on the validation set
#represents how well the model performs on data it hasn't seen during training
```

```
#Mapping integers back to labels for prediction
class_names = ['negative', 'neutral', 'positive']

def predict_sentiment(text):

    if not text or not isinstance(text, str):
        return "invalid input", 0.0

    model.eval() #eval mode

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    #tokenizing the input text
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128) #If the input text exceeds the maximum length, it will be truncated to the specified length
    #If the input text is shorter than the max length, it will be padded with zeros to ensure all inputs have the same length
    inputs = {key: val.to(device) for key, val in inputs.items()}


    #Getting the model's prediction without gradients for efficiency
    #When you don’t need to update the model’s parameters, you wrap your code inside this context to prevent the computation of gradients,
    #which saves memory and speeds up computations.
    with torch.no_grad():
        outputs = model(**inputs) #forwards pass through the model to compute logits
        logits = outputs.logits #gets the raw logits from the model
        probs = torch.nn.functional.softmax(logits, dim=1) #converts logits into probs
        predicted_class_id = logits.argmax().item()  #gets the index of the max logit
        confidence = probs[0][predicted_class_id].item()  #gets the confidence


    return class_names[predicted_class_id], confidence


#During predicting, we don’t need to compute gradients

#Logits are the raw output scores (before applying activation functions like softmax)

#Softmax: Converts the logits into probabilities, so each class gets a value between 0 and 1, which represents the probability of that class being correct
```

```
#performs K-means clustering on embeddings and reduces dimensions to 2D for visualization
#embedding: high dimensional vector representations of texts

def perform_clustering(embeddings, n_clusters=8, random_state=42):
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
    cluster_labels = kmeans.fit_predict(embeddings) #assigning each point to a cluster

    #Dimensionality reduction for visualization
    reducer = umap.UMAP(n_components=2, random_state=random_state)
    reduced_embeds = reducer.fit_transform(embeddings)

    return cluster_labels, reduced_embeds

#labels are numeric here

#PCA: Linear, fast, good for linearly separable data
#t-SNE: Non-linear, good for visualization, computationally expensive
#UMAP: Non-linear, scalable, preserves local and global structure
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
#Getting keywords for a cluster using different methods
#Using BERT embeddings to find words most similar to the document context

def get_cluster_keywords(texts, cluster_num, n_keywords=5, method='keybert'): #keyword extraction
    if not texts or len(texts) == 0: #empty clusters
        return ["no_text_data"]

    #KeyBERT uses BERT embeddings to find most similar words to the doc
    try:
        if method == 'keybert':
            #Joining all texts for better keyword extraction
            combined_text = " ".join(texts)
            keywords = kw_model.extract_keywords(
                combined_text,
                keyphrase_ngram_range=(1, 2), #Allowing 1-2 word phrases, capturing multi-word terms
                stop_words='english',
                top_n=n_keywords,
                use_mmr=True,  #Maximal marginal relevance for diversity, prevents redundant keywords
                diversity=0.5  #Balance between relevance and diversity
            )

            #extract keywords even if empty
            return [kw[0] for kw in keywords] if keywords else ["no_keywords_found"]

        elif method == 'tfidf':
            #Using TF-IDF to find words with highest term freq-inverse doc freq
            vectorizer = TfidfVectorizer(stop_words='english', max_features=1000) #vocab size
            X = vectorizer.fit_transform(texts) #creating document - term matrix
            mean_tfidf = X.mean(axis=0).A1 #avg scores per word
            top_indices = mean_tfidf.argsort()[::-1][:n_keywords] #getting top indices, reversing the array, slicing with -1 step
            return vectorizer.get_feature_names_out()[top_indices].tolist() #returns the vocab in the same order

        #.A1 converts the sparse TF-IDF matrix into a flat NumPy array for efficient sorting of importance scores
        #downweigthing common words with IDF, highlighting distinctive words for each cluster

        elif method == 'lda':
            #probabilistic topic model
            vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2)
            dtm = vectorizer.fit_transform(texts) #doc-term matrix
            if dtm.shape[0] < 2: #num of docs in the cluster
                return ["<not enough data>"]
            lda = LatentDirichletAllocation(n_components=1, random_state=42) #Forces LDA to find just one topic for this cluster
            lda.fit(dtm) #learning topic distribution
            words = vectorizer.get_feature_names_out()
            return [words[i] for i in lda.components_[0].argsort()[-n_keywords:][::-1]]

            #LDA requires multiple documents to find meaningful topics
            #Can't calculate word co-occurrence patterns without multiple docs
            #max_df discards words appearing in > 95% of cluster documents, removes stopwords that survived preprocessing
            #min_df ignores words that appear in < 2 documents, eliminates noise from rare terms

    except Exception as e:
        print(f"Error extracting keywords for cluster {cluster_num}: {str(e)}")
        return ["keyword_extraction_failed"]
```

```
#combining clustering, keyword extraction and analysis
def analyze_clusters(df, embeddings, n_clusters=8):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    df['cluster'] = kmeans.fit_predict(embeddings)

    #UMAP handles high-dimensional embeddings well
    #visualization
    reducer = umap.UMAP(n_components=2, random_state=42)
    reduced_embeds = reducer.fit_transform(embeddings)

    #cluster statistics with fallback for empty clusters
    cluster_stats = []
    for i in range(n_clusters):
        cluster_texts = df[df['cluster'] == i]['clean_tweet'].tolist()

        #Try KeyBERT first, fallback to TF-IDF if it fails
        keywords = get_cluster_keywords(cluster_texts, i, method='keybert')
        if keywords == ["no_keywords_found"] or keywords == ["keyword_extraction_failed"]:
            keywords = get_cluster_keywords(cluster_texts, i, method='tfidf')

        cluster_stats.append({
            'cluster_num': i,
            'count': len(cluster_texts),
            'keywords': keywords,
            'sample_tweets': df[df['cluster'] == i]['tweet'].sample(min(2, len(cluster_texts))).tolist() #only rows in the curr cluster
            #how many tweets to sample from the cluster
            #if the cluster is too small, it doesn't give an error

        })

    return cluster_stats, reduced_embeds #dictionary with cluster characteristics and 2D projections


n_clusters = 8
cluster_stats, reduced_embeds = analyze_clusters(df, embeddings, n_clusters)
```

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
#Extracts topic keywords using LDA for a given cluster

def get_lda_topic_keywords(cluster_num, n_keywords=5):
    cluster_tweets = df[df['cluster'] == cluster_num]['clean_tweet'] #gets clean tweets for the specified cluster
    if cluster_tweets.empty:
        return ["<no data>"]

    vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2)
    dtm = vectorizer.fit_transform(cluster_tweets) #document term matrix
    if dtm.shape[0] < 2: #Number of documents in the cluster
        #LDA requires multiple documents to identify co-occurrence patterns
        return ["<not enough data>"]

    lda = LatentDirichletAllocation(n_components=1, random_state=42) #Forces single topic per cluster
    lda.fit(dtm)
    words = vectorizer.get_feature_names_out()
    topic_keywords = [words[i] for i in lda.components_[0].argsort()[-n_keywords:][::-1]] #Selects top N indices and reverses order
    return topic_keywords

#lda.components_[0] gets word distribution for the single topic

print("\n=== LDA Topic Keywords per Cluster ===")
for i in range(n_clusters):
    lda_keywords = get_lda_topic_keywords(i)
    print(f"Cluster {i}: {', '.join(lda_keywords)}")
```

- LDA identifies latent topics via word co-occurrence

- TF-IDF highlights distinctive terms statistically

Latent Dirichlet Allocation (LDA) is a probabilistic model used for topic modeling in NLP. It automatically discovers hidden (latent) thematic patterns in a collection of documents by analyzing word co-occurrence patterns.
- Identifies abstract topics in a set of documents

- Handles words with multiple meanings

- Captures semantic relationships between words

- Each topic is represented as a distribution over words

- Each document is represented as a distribution over topics

- Assumes documents are generated by randomly mixing topics

- Topics are discovered by "reverse-engineering" this process

```
%%writefile app.py
from fastapi import FastAPI, Request, Form
from fastapi.responses import HTMLResponse
from pydantic import BaseModel
from transformers import BertTokenizer, BertForSequenceClassification
import torch
import torch.nn.functional as F

#Load tokenizer and model
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=3)
model.load_state_dict(torch.load("bert_sentiment_weights.pt", map_location=torch.device("cpu")))
model.eval()

class_names = ['negative', 'neutral', 'positive']
app = FastAPI()

html = """
<!DOCTYPE html>
<html>
<head><title>Sentiment Analysis</title></head>
<body>
  <h2>Enter a sentence:</h2>
  <form action="/predict/" method="post">
    <input name="text" type="text" size="50">
    <input type="submit">
  </form>
</body>
</html>
"""

@app.get("/", response_class=HTMLResponse)
async def root():
    return html

@app.post("/predict/", response_class=HTMLResponse)
async def predict_sentiment(text: str = Form(...)):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        probs = F.softmax(logits, dim=1)
        predicted_class_id = logits.argmax().item()
        confidence = probs[0][predicted_class_id].item()

    return f"""
    <html>
        <body>
            <h2>Sentiment:</h2>
            <p><strong>Text:</strong> {text}</p>
            <p><strong>Predicted:</strong> {class_names[predicted_class_id]}</p>
            <p><strong>Confidence:</strong> {confidence:.4f}</p>
            <a href="/">Try another</a>
        </body>
    </html>
    """
```
